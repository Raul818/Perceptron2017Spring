```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import cv2
```

```python
# Load an Image
im = cv2.imread('./im.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
im = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)

plt.figure(0)
plt.axis("off")
plt.imshow(im, cmap='gray')
plt.show()
```
